Data Loading

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, precision_recall_curve, accuracy_score
from imblearn.over_sampling import SMOTE
import numpy as np
from keras.layers import InputLayer, Dense, BatchNormalization
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
import tensorflow as tf
from warnings import filterwarnings
filterwarnings('ignore')
df = pd.read_csv('dataset/creditcard.csv')
df
```

Feature Scaling

```python
scaler = RobustScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
df['Amount'].hist()
df['Amount'].describe()

# No outlier, so we can Standardize Time column
time = df['Time']
df['Time'] = (time - time.min()) / (time.max() - time.min())
```

```python
df.describe()
```

Handling Imbalanced Dataset using SMOTE

```python
X = df.drop('Class', axis=1)
y = df['Class']

# Apply SMOTE to the entire dataset
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
print(pd.Series(y_smote).value_counts())
```

K-Folds Cross Validation

```python
k = 10
kf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
```

Logistic Regression Model

```python
logistic_accuracies = []

for train_index, test_index in kf.split(X_smote, y_smote):
    X_train_k, X_test_k = X_smote.iloc[train_index], X_smote.iloc[test_index]
    y_train_k, y_test_k = y_smote.iloc[train_index], y_smote.iloc[test_index]

    logistic_model = LogisticRegression(max_iter = 1000)
    logistic_model.fit(X_train_k, y_train_k)

    y_pred_k = logistic_model.predict(X_test_k)
    accuracy = accuracy_score(y_test_k, y_pred_k)
    logistic_accuracies.append(accuracy)

print(f'Logistic Regression accuracies for each fold: {logistic_accuracies}')
print(f'Average accuracy: {np.mean(logistic_accuracies)}')
```

ROC Curve of the Logistic Regression

```python
y_pred_proba = logistic_model.predict_proba(X_smote)[:, 1]  # Changed X_test_k to X_smote for consistency
roc_auc = roc_auc_score(y_smote, y_pred_proba)  # Changed y_test_k to y_smote for consistency
print(f'ROC-AUC Score: {roc_auc}')

fpr, tpr, thresholds = roc_curve(y_smote, y_pred_proba)  # Changed y_test_k to y_smote for consistency
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_smote, y_pred_proba)  # Changed y_test_k to y_smote for consistency
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```

```python
y_pred = logistic_model.predict(X_smote)  # Changed X_test_k to X_smote for consistency
adjusted_report = classification_report(y_smote, y_pred)  # Changed y_test_k to y_smote for consistency
print(f'Classification Report:\n', adjusted_report)
```

```python
def neural_network_predictions(model, X):
    return (model.predict(X).flatten() >= 0.5).astype(int)
```

Shallow Neural Network Model

```python
shallow_nn_accuracies = []

for train_index, test_index in kf.split(X_smote, y_smote):
    X_train_k, X_test_k = X_smote.iloc[train_index], X_smote.iloc[test_index]
    y_train_k, y_test_k = y_smote.iloc[train_index], y_smote.iloc[test_index]

    #Define the neural network model
    shallow_nn = Sequential()
    shallow_nn.add(InputLayer((X_train_k.shape[1],)))
    shallow_nn.add(Dense(2, activation="relu"))
    shallow_nn.add(BatchNormalization())
    shallow_nn.add(Dense(1, activation="sigmoid"))

    checkpoint = ModelCheckpoint('shallow_nn_model.keras', save_best_only=True)
    shallow_nn.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
    shallow_nn.summary()

    #Train the model
    shallow_nn.fit(X_train_k, y_train_k, validation_data=(X_test_k, y_test_k), epochs=10, callbacks=[checkpoint])

    #Predict and evaluate
    y_pred_nn_k = neural_network_predictions(shallow_nn, X_test_k)
    accuracy_nn = accuracy_score(y_test_k, y_pred_nn_k)
    shallow_nn_accuracies.append(accuracy_nn)

print(f'Shallow Neural Network accuracies for each fold: {shallow_nn_accuracies}')
print(f'Average accuracy: {np.mean(shallow_nn_accuracies)}')
```

```python
shallow_nn.summary()
```

ROC-AUC Score and ROC Curve for Shallow Neural Network

```python
y_pred_nn = neural_network_predictions(shallow_nn, X_smote) # Changed X_test_k to X_smote for consistency
roc_auc_nn = roc_auc_score(y_smote, y_pred_nn)  # Changed y_test_k to y_smote for consistency
print(f"ROC-AUC Score: {roc_auc_nn}")
print(f"ROC-AUC Score: {roc_auc_nn}")

fpr, tpr, thresholds = roc_curve(y_smote, y_pred_nn) # Changed y_test_k to y_smote for consistency
plt.figure(figsize=(8, 6))
plt.figure(figsize = (8,6))
plt.plot(fpr, tpr, label =f'ROC Curve (area = {roc_auc_nn:.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC Curve")
plt.legend(loc = "best")
plt.show()
```

Precision and Recall for Shallow Neural Network

```python
precision, recall, thresholds = precision_recall_curve(y_smote, y_pred_nn) # Changed y_test_k to y_smote for consistency
plt.figure(figsize = (8, 6))
plt.plot(precision, recall, marker ='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```

Best Threshold for Shallow Neural Network

```python
best_threshold_index_nn = np.argmax(precision * recall)
best_threshold_nn = thresholds[best_threshold_index_nn]
print(thresholds)
print(f'Best Threshold: {best_threshold_nn}')

y_pred_nn_new = (y_pred_nn >= best_threshold_nn).astype(int)
```

Classification Report for Shallow Neural Network Model

```python
print(classification_report(y_smote, y_pred_nn_new))
```

Random Forest Classifier with K-Fold Cross-Validation

```python
rfc_accuracies = []

for train_index, test_index in kf.split(X_smote, y_smote):
    X_train_k, X_test_k = X_smote.iloc[train_index], X_smote.iloc[test_index]
    y_train_k, y_test_k = y_smote.iloc[train_index], y_smote.iloc[test_index]

    # Train the model
    rfc_model = RandomForestClassifier(max_depth = 2, n_jobs = -1, random_state=42)
    rfc_model.fit(X_train_k, y_train_k)

    # Predict and evaluate
    y_pred_k = rfc_model.predict(X_test_k)
    accuracy = accuracy_score(y_test_k, y_pred_k)
    rfc_accuracies.append(accuracy)
    
print(f'RandomForestClassifier accuracies for each fold: {rfc_accuracies}')
print(f'Average accuracy: {np.mean(rfc_accuracies):.2f}')
```
